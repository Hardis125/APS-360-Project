```python
import os
import shutil
import random

# Define the input and output folders
input_folder = '/content/drive/My Drive/APS360Project/Hardis_Data'
output_folder = '/content/drive/My Drive/APS360Project/split'

# Create output folders if they don't exist
os.makedirs(output_folder, exist_ok=True)
os.makedirs(os.path.join(output_folder, 'train'), exist_ok=True)
os.makedirs(os.path.join(output_folder, 'val'), exist_ok=True)
os.makedirs(os.path.join(output_folder, 'test'), exist_ok=True)

# Set the split ratios for dataset (70% train, 15% validation, 15% test)
split_ratios = (0.7, 0.15, 0.15)

# Get a list of all image files in the input folder
image_files = [filename for filename in os.listdir(input_folder) if any(filename.endswith(extension) for extension in ['.jpg', '.jpeg', '.png'])]

# Shuffle the list of image files
random.shuffle(image_files)

# Calculate split sizes
num_files = len(image_files)
num_train = int(split_ratios[0] * num_files)
num_val = int(split_ratios[1] * num_files)
num_test = num_files - num_train - num_val

# Split the image files
train_files = image_files[:num_train]
val_files = image_files[num_train:num_train + num_val]
test_files = image_files[num_train + num_val:]

# Copy files to output folders
def copy_files(files, folder):
    for file in files:
        shutil.copy(os.path.join(input_folder, file), os.path.join(output_folder, folder, file))

copy_files(train_files, 'train')
copy_files(val_files, 'val')
copy_files(test_files, 'test')

print("Dataset split successfully.")
```

```python
import os
import shutil

def remove_empty_subfolders(input_folder):
    for root, dirs, files in os.walk(input_folder, topdown=False):
        for dir_name in dirs:
            dir_path = os.path.join(root, dir_name)
            if len(os.listdir(dir_path)) == 0:  # Check if the folder is empty
                print(f"Removing empty folder: {dir_path}")
                os.rmdir(dir_path)

def split_dataset(input_folder, output_folder, train_ratio=0.7, val_ratio=0.15, test_ratio=0.15):
    remove_empty_subfolders(input_folder)  # Remove empty subfolders first

    for root, dirs, files in os.walk(input_folder):
        num_files = len(files)
        train_split = int(train_ratio * num_files)
        val_split = int(val_ratio * num_files)

        for i, file_name in enumerate(files):
            source_file_path = os.path.join(root, file_name)
            if i < train_split:
                destination_folder = "train"
            elif i < train_split + val_split:
                destination_folder = "val"
            else:
                destination_folder = "test"

            destination_folder_path = os.path.join(output_folder, destination_folder)
            os.makedirs(destination_folder_path, exist_ok=True)
            destination_file_path = os.path.join(destination_folder_path, file_name)

            print(f"Copying {source_file_path} to {destination_file_path}")
            shutil.copy(source_file_path, destination_file_path)

# Usage
input_folder = "/content/drive/My Drive/APS360Project/ProjectImg"
output_folder = "/content/drive/My Drive/APS360Project/split"
split_dataset(input_folder, output_folder)
```
